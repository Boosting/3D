```python
from landmarkPredict import predictImage 
import numpy as np

point=predictImage('wufan.txt')
```

```python
import os

# save the 2D landmarks
print point.shape
f=open('wufan.pts','wb')
f.writelines('version: 1\n')
f.writelines('n_points:  68\n')
f.writelines('{\n')

for i in range(point.shape[1]/2):
    x=point[0][2*i]
    y=point[0][2*i+1]
    line='%f %f\n' % (x, y)
    f.writelines(line)
f.writelines('}')
f.close()
```

```python
#help(eos.morphablemodel.PcaModel)
def plywrite(s,t):
    nver=s.shape[0]/3
    nface=1
    f=open('test.ply','wb')
    f.writelines('ply\n')
    f.writelines('format ascii 1.0\n')
    f.writelines('comment made by YadiraF\n')
    f.writelines('element vertex %d\n' % (nver))
    f.writelines('property float x\n')
    f.writelines('property float y\n')
    f.writelines('property float z\n')
    f.writelines('property uchar red\n');
    f.writelines('property uchar green\n');
    f.writelines('property uchar blue\n');
    f.writelines('element face %d\n' % (nface));
    f.writelines('property list uchar int vertex_indices\n');
    f.writelines('end_header\n');
    
    # writing 3D shape and vertex color
    for i in range(nver):
        f.writelines('%f %f %f\n' % (s[i*3],s[i*3+1],s[i*3+2]))
    for i in range(nver):
        f.writelines('%d %d %d\n' % (t[i*3],t[i*3+1],t[i*3+2]))
    f.close()
    
t=np.array(s,dtype=np.uint8)
print s
print t

plywrite(s,t)
```
