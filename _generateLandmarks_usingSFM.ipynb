Generate Landmarks using Surrey Face Model

```python
'''
This ipynb file is used for debugging and visualization.
The completed file is generateLandmarks.py in the same dir.
'''
import numpy as np
import sys
sys.path.append('./eos/install/bin')
import eos

model = eos.morphablemodel.load_model("3D-model/sfm_shape_3448.bin")
import matplotlib.pyplot as plt
import scipy.io as sio
% matplotlib inline

def read_txt(name):
    fp=open(name,'r')
    line=fp.readlines()
    fp.close()
    points=np.array(line,dtype=int)
    return points

points=read_txt("3D-model/50_3D_point.txt")


mu=2*np.random.random_sample()-1 #[-1,1]
sigma=np.random.random_sample()/2 #[0,0.5]
param = np.random.normal(mu, sigma, 63)
s = model.get_shape_model().draw_sample(param)
shape = np.array(s)

# shape and 3D landmark points
print len(shape)/3
shape=np.reshape(shape,(len(shape)/3,3)).T
#
shape=np.vstack((shape,np.ones((1,shape.shape[1]))))
print shape.shape

# 3x50-->4x50
vertices= shape[:,points]
print vertices.shape


# pinhole camera model
#http://homepages.inf.ed.ac.uk/rbf/CVonline/LOCAL_COPIES/EPSRC_SSAZ/node3.html
#http://docs.opencv.org/2.4/modules/calib3d/doc/camera_calibration_and_3d_reconstruction.html
#http://www.cs.ucf.edu/~mtappen/cap5415/lecs/lec19.pdf


# https://zh.wikipedia.org/wiki/%E6%97%8B%E8%BD%AC%E7%9F%A9%E9%98%B5
# r,p,y in [0,pi]
def get_rotation_matrix(r,p,y):

    roll=np.array([[1,         0,       0],
                   [0, np.cos(r), -np.sin(r)],
                   [0, np.sin(r),  np.cos(r)]])
    pitch=np.array([[np.cos(p),  0, np.sin(p)],
                    [0,          1,         0],
                    [-np.sin(p), 0, np.cos(p)]])
    yaw=np.array([[np.cos(y), -np.sin(y), 0],
                  [np.sin(y),  np.cos(y), 0],
                  [        0,          0, 1]])
    
    R=yaw.dot(pitch.dot(roll))
    print 'R',R
    return R
    
    
# x=P*X P=C*[R|T]=intr*extr
def get_projection_matrix(p):
    [pitch,yaw,roll,tx,ty,tz,alpha_u,alpha_v,u_0,v_0]= p
    
    intr=np.array([[alpha_u,     0.0,   u_0],
                   [    0.0, alpha_v,   v_0],
                   [    0.0,     0.0,   1.0]])
    
    R=get_rotation_matrix(roll,pitch,yaw) # roll pitch yaw 
    
    T=np.array([[tx],[ty],[tz]]) # tx ty tz   
    
    extr=np.hstack((R,T))

    P=intr.dot(extr)
    return P

p=[np.pi/4,np.pi/6,np.pi/8, 20,0,0, 1,1, 0,0] #r p y
P=get_projection_matrix(p)

s2=P.dot(shape)
l2=P.dot(vertices)
m=P.dot(shape[:,270])
#plt.plot(s2[0],s2[1],'b.')
plt.plot(s2[0,:],s2[1,:],'b+')
plt.plot(l2[0],l2[1],'r.')

plt.axis([-150, 150, -150, 150])
print m

# center
center_ind=17
#plt.plot(l2[0,center_ind],l2[1,center_ind],'ko')

# normalize
ln=l2.T-m.T
ln=ln.T
print ln.shape
#plt.plot(ln[0],ln[1],'g.')

# 50->9
print points[center_ind]
# 9 ind
#[1 2 3 4 5 6 7 8 9] in 9 landmarks
#[34  37  40  43  46  49  55  63  67] in 68 landmarks
#[270 177 181 614 610 398 329 812 411]in 3D vertices
#[17, 20, 23, 26, 29, 32, 35, 38, 41] 9 points index in 50 landmarks
ind=read_txt("3D-model/9_index.txt")
landmarks=l2[:2,ind]
print landmarks[:,0]
tmp=landmarks.T-landmarks[:,0].T
landmarks=tmp.T
print points[ind]
# d 3-4
p1=landmarks[:,2]
p2=landmarks[:,3]
print p1
lamb= np.linalg.norm(p1-p2)

landmarks/=lamb

plt.figure()
plt.plot(landmarks[0,:],landmarks[1,:],'bo')

```
